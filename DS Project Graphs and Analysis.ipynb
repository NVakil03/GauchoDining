```python
def min_max_items(df):

    df['Served'] = pd.to_numeric(df['Served'], errors='coerce')

    # Find index of min and max
    min_idx = df['Served'].idxmin()
    max_idx = df['Served'].idxmax()

    # Get the rows
    min_row = df.loc[min_idx]
    max_row = df.loc[max_idx]

    return min_row, max_row

min_max_items(data)
```

```python
#separate dataframe for just protein breakfast items, without the serving temperature because that's kind of irrelevant?
proteins = pd.read_csv('Proteins.csv')
update = proteins.drop('Serving Temperature (°F)', axis = 1)
update
```

```python
# Filter for protein-rich breakfast items
#there's a large gap between jan and feb because of missing data, which we hope to fill 
#most items seemed to be increasing at some point leading up to finals week and then decreasing at the end of week 10

import matplotlib.pyplot as plt
protein_keywords = ["Scrambled Eggs", "Egg Whites", "Sausage", "Frittata"]

# Filter for breakfast only and items containing protein keywords
protein_breakfast_df = proteins[proteins['Meal'].str.lower().isin(['breakfast', 'brunch']) &
proteins['Food Item'].str.contains('|'.join(protein_keywords), case=False, na=False)].copy()

# Convert 'Day' to datetime and clean 'Served'
protein_breakfast_df['Day'] = pd.to_datetime(protein_breakfast_df['Day'], errors='coerce')
protein_breakfast_df['Served'] = pd.to_numeric(protein_breakfast_df['Served'], errors='coerce')

# Group by Day and Food Item
grouped_proteins = protein_breakfast_df.groupby(['Day', 'Food Item'])['Served'].sum().unstack().fillna(0)

# Plot
grouped_proteins.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Protein Items Served During Breakfast/Brunch Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity Served')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#function to describe the above protein data
#chicken sausage patty and scrambled eggs are the two highest, which makes sense because people tend to ask for them together

def describe_protein_servings(dataframe):

    protein_keywords = ["Scrambled Eggs", "Egg Whites", "Sausage", "Frittata"]
    df = dataframe[
        dataframe['Meal'].str.lower().isin(['breakfast', 'brunch']) &
        dataframe['Food Item'].str.contains('|'.join(protein_keywords), case=False, na=False)
    ].copy()

    df['Food'] = (
        df['Food Item']
        .str.replace('-', ' ', regex=False)
        .str.lower()
        .str.replace(r'\s+', ' ', regex=True)
        .str.strip()
    )

    df['Served'] = pd.to_numeric(df['Served'], errors='coerce')
    description = df.groupby('Food')['Served'].describe()
    return description

describe_protein_servings(proteins)
```
